## Libraries

```python
import pandas as pd # library for data manipulation
import matplotlib.pyplot as plt # library for visualization
import seaborn as sns # library for visualization
sns.set() # this command sets the seaborn chart style as the default
```

## Stacked Area Chart

```python
df_fuel_engine_types = pd.read_csv("stacked_area_chart_data.csv")
df_fuel_engine_types
```

```python
#Raw chart 

plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Diesel"],
             df_fuel_engine_types["Petrol"],
             df_fuel_engine_types["Gas"])      
             
plt.show()
             
             
        
```

```python
#Format chart 

colors_list = ["#011638", "#7e2987", "#ef2026"]
labels_name = ["Diesel", "Petrol", "Gas"]
sns.set_style("white")

plt.figure(figsize = (12, 6))

plt.stackplot(df_fuel_engine_types["Year"],
              df_fuel_engine_types["Diesel"],
             df_fuel_engine_types["Petrol"],
             df_fuel_engine_types["Gas"],
             colors = colors_list,
             edgecolor = 'none')      


plt.xticks(df_fuel_engine_types["Year"], rotation = 45)
plt.legend(labels = labels_name, loc = "upper left")
plt.ylabel("Number of Cars", fontsize = 13)
plt.title("Popularity of Engine Fuel Types (1982 - 2016)", fontsize = 13, weight = "bold")
sns.despine()

plt.tight_layout()
plt.savefig("Stacked-Area_Chart.png")
plt.show()
```
