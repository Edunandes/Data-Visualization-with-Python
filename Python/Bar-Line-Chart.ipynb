## Libraries

```python
import pandas as pd # library for data manipulation
import matplotlib.pyplot as plt # library for visualization

from matplotlib.ticker import PercentFormatter
```

## Bar and Line Chart

```python
df_kdnuggets = pd.read_csv("bar_line_chart_data.csv")
df_kdnuggets
```

```python
fig, ax = plt.subplots(figsize = (10, 7))

#bar plot
ax.bar(df_kdnuggets["Year"],
      df_kdnuggets["Participants"],
      color = 'k')

ax.set_ylabel("Number of Participants", weight = "bold")
ax.tick_params(axis = "y", width = 2, labelsize = "large")


#twin axes for the line plot 
ax1 = ax.twinx()
ax1.set_ylim(0,1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))

#Line plot 
ax1.plot(df_kdnuggets["Year"],
        df_kdnuggets["Python Users"],
        color = "#b60000", 
        marker = "D")

ax1.set_ylabel("Python Users", color = "#b60000", weight = "bold")
ax1.tick_params(axis = "y", colors = "#b60000", width = 2, labelsize = "large")

#set the title 
ax.set_title("KD Nuggets Survey Python Users (2012 - 2019)", fontsize = "14", weight = "bold")

#set background color 
fig.patch.set_facecolor('white')
ax.set_facecolor('white')

#border around the chart area
for spine in ax1.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1.5)


plt.tight_layout()
plt.savefig("Bar-Line-Chart.png")
plt.show()
```
