## Libraries

```python
import pandas as pd # library for data manipulation
import matplotlib.pyplot as plt # library for visualization
import seaborn as sns # library for visualization
sns.set() # this command sets the seaborn chart style as the default
```

# Histogram Chart

```python
df_real_estate = pd.read_csv("histogram_data.csv")
df_real_estate
```

```python
#raw histogram chart 

plt.hist(df_real_estate["Price"])

plt.show()

#10 bins
```

```python
sns.set_style("white")
plt.figure(figsize = (8,6))


plt.hist(df_real_estate["Price"],
        bins = 8,
        color = "#108A99")

plt.title("Distribution of Real Estate Prices", fontsize = 14, weight = "bold")
plt.xlabel("Price in (000' $)")
plt.ylabel("Number of Properties")

sns.despine()

plt.tight_layout()
plt.savefig("Histogram-Chart-Real-Estate.png")
plt.show()
```
