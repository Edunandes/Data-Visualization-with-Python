## Libraries

```python
import pandas as pd # library for data manipulation
import matplotlib.pyplot as plt # library for visualization
import seaborn as sns # library for visualization
sns.set() # this command sets the seaborn chart style as the default
```

## Regression Plot

```python
df_ad_expenditure = pd.read_csv("regression_data.csv")
df_ad_expenditure
```

```python
#raw chart 

sns.regplot(x = "Budget",
           y = "Sales",
           data = df_ad_expenditure)

plt.show()
```

```python
#sns.regplot()

sns.set(rc = {'figure.figsize' : (9,6)})
sns.regplot(x = "Budget",
           y = "Sales",
           data = df_ad_expenditure,
           scatter_kws = {'color': 'k'},
           line_kws = {'color':'red'})


plt.xlabel("Ad Expenditure in (000`s $)")
plt.ylabel("Sales in (000`s $)")
plt.title("Effect of Ad Expenditure on Sales", fontsize = 14, weight = "bold")

plt.tight_layout()
plt.savefig("Regression-Plot.png")
plt.show()
```

```python
#sns.lmplot()
sns.lmplot( x = "Budget",
          y = "Sales",
          data = df_ad_expenditure, 
          height = 10,
           scatter_kws = {'color': 'k'},
           line_kws = {'color':'red'})

plt.xlabel("Ad Expenditure in (000`s $)")
plt.ylabel("Sales in (000`s $)")
plt.title("Effect of Ad Expenditure on Sales", fontsize = 14, weight = "bold")

plt.tight_layout()

plt.show()
```
